```python
import random as r
```

```python
secrets = ['mango','orange','banana','melon','apple']
guessed = r.choice(secrets)
gusseDash = ['-']*len(guessed)
chances = len(guessed)+2
while chances>0:
    usin = input('Enter the letter you gussed: ')
    if len(usin)>1 or usin.isdecimal():
        print('Invalid Entry')
        continue
    else:
        usin=usin.lower()
        coutr = 0
        if usin in guessed:
            for i in range(len(guessed)):
                if guessed[i]==usin and gusseDash[i]=='-':
                    gusseDash[i]=usin
                    coutr=coutr+1
                else:
                    continue
            if coutr>0:
               hword= ''.join(gusseDash)
               print(hword)
               if hword == guessed:
                   print('You gussed right')
                   break
            else:
                print('You may have entered this letter before')
        else:
            print('Wrong guess')
    chances=chances-1
    print('Remaining chances are: ',chances)

```
